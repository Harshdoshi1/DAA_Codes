```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

image_path = 'cheetah.jpg'

def apply_convolution_filter(image_path, mask):
    print(f"Attempting to load image from: {image_path}")
    print(f"File exists: {os.path.exists(image_path)}") 
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Error: Could not load image! Check the path and file format.")
        return
    
    kernel = np.array(mask, dtype=np.float32)
    filtered_image = cv2.filter2D(image, -1, kernel)
    
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.title("Filtered Image")
    plt.imshow(filtered_image, cmap='gray')
    plt.axis('off')
    
    cv2.imwrite('filtered_image.png', filtered_image)
    
    plt.tight_layout()
    plt.show()

mask = [
    [-1, -1, -1],
    [-1, 20, -1],
    [-1, -1, -1]
]

apply_convolution_filter(image_path, mask)
```
